```python
y=df['Diabetes_binary'].to_numpy()
x = df.drop(columns=['Diabetes_binary']).to_numpy()

x
y
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
xtrain.shape, xtest.shape
```

```python
ytrain.shape, ytest.shape
```

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score, accuracy_score,roc_curve,auc
```

```python
import numpy as np
```

```python

testing_error_rate = [] #array of error rates generated from testing
testing_f1_scores = [] #array of F1 scores from testing
testing_accuracies = [] #array of accuracies from testing

roc_curve_scores = [] #testing ROC values

training_error_rate = [] #array of training error rates
training_f1_scores = [] #array of training f1 scores
training_accuracies = [] #array of training accuracies

f1_testing_k = 0; # value of k when F1 is highest in testing
a_testing_k =0; #value of k when accuracy is highest in testing 

f1_training_k = 0; #value of k when f1 is highest in training
a_training_k =0; #value of k when accuracy is highest in training
 
maxk_f1_test =0; #maximum value of f1 scores in testing
maxk_accuracy_test = 0; #maximum value of accuracy in testing

maxk_f1_train =0; #maximum value of f1 scores in training
maxk_accuracy_train = 0; #maximum value of accuracy in training

roc_auc=0;

for i in range(1,60): #i is the value of K which the hyperparameter getting tuned.
    
    knn_training = KNeighborsClassifier(n_neighbors=i)
    knn_training.fit(xtrain,ytrain)
    pred_i_training = knn_training.predict(xtrain)
    training_error_rate.append(np.mean(pred_i_training !=ytrain))
    
    confusion_matrix(ytrain,pred_i_training)
    
    tn_train, fp_train, fn_train, tp_train = confusion_matrix([0, 1, 0, 1], [1, 1, 1, 0]).ravel()
    
   
    training_accuracy = accuracy_score(ytrain,pred_i_training) #finds the training accuracy for a particular i value
    training_accuracies.append(training_accuracy) 
    f1_score1_training = f1_score(ytrain,pred_i_training) #finds the training f1 score 
    training_f1_scores.append(f1_score1_training)
    if f1_training_k < f1_score1_training: #finds the maximum training f1
        maxk_f1_train = i
        f1_training_k = f1_score1_training
        
    if a_training_k<training_accuracy: #finds maximum training accuracy
        maxk_accuracy_train = i
        a_training_k = training_accuracy
        
    #--------------------------------------------------------------------------------------------------------------
    
    knn_testing = KNeighborsClassifier(n_neighbors=i)
    knn_testing.fit(xtrain,ytrain)
    pred_i = knn_testing.predict(xtest)
    testing_error_rate.append(np.mean(pred_i !=ytest))
    
    confusion_matrix(ytest,pred_i)
    
    tn, fp, fn, tp = confusion_matrix([0, 1, 0, 1], [1, 1, 1, 0]).ravel()
    
   
    testing_accuracy = accuracy_score(ytest,pred_i) #finds the testing accuracy
    testing_accuracies.append(testing_accuracy)
    f1_score1_testing = f1_score(ytest,pred_i) #finds the testing f1 score
    testing_f1_scores.append(f1_score1_testing)
    if f1_testing_k < f1_score1_testing: #gets the maximum testing f1's K value
        maxk_f1_test = i
        f1_testing_k = f1_score1_testing
        
    if a_testing_k < testing_accuracy: #gets the maximum accuracy's K value
        maxk_accuracy_test = i
        a_testing_k = testing_accuracy
        
    # Calculate ROC curve
    fpr, tpr, _ = roc_curve(ytest, pred_i)
    roc_auc = auc(fpr, tpr)
    roc_curve_scores.append(roc_auc)

    
print("Maximum testing F1 score is when k = ",maxk_f1_test);
print("Maximum testing F1 score is ",f1_testing_k)
print("Maximum testing Accuracy is when k = ",maxk_accuracy_test)
print("Maximum testing Accuracy is ",a_testing_k)

print("Maximum training F1 score is when k = ",maxk_f1_train)
print("Maximum training F1 score is ",f1_training_k)
print("Maximum training Accuracy is when k = ",maxk_accuracy_train)
print("Maximum training Accuracy is ",a_training_k)
```
