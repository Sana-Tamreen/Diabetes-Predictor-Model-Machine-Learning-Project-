```python
import pandas as pd
df = pd.read_csv("Diabetes.csv")
```

```python
y=df['Diabetes_binary'].to_numpy()
x = df.drop(columns=['Diabetes_binary']).to_numpy()

x
y
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
xtrain.shape, xtest.shape
```

```python
ytrain.shape, ytest.shape
```

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score, accuracy_score,roc_curve,auc
```

```python
import numpy as np
```

```python

testing_error_rate = []
testing_f1_scores = []
testing_accuracies = []

roc_curve_scores = []

training_error_rate = []
training_f1_scores = []
training_accuracies = []

f1_testing_k = 0;
a_testing_k =0;

f1_training_k = 0;
a_training_k =0;

maxk_f1_test =0;
maxk_accuracy_test = 0;

maxk_f1_train =0;
maxk_accuracy_train = 0;

roc_auc=0;

for i in range(1,60):
    
    knn_training = KNeighborsClassifier(n_neighbors=i)
    knn_training.fit(xtrain,ytrain)
    pred_i_training = knn_training.predict(xtrain)
    training_error_rate.append(np.mean(pred_i_training !=ytrain))
    
    confusion_matrix(ytrain,pred_i_training)
    
    tn_train, fp_train, fn_train, tp_train = confusion_matrix([0, 1, 0, 1], [1, 1, 1, 0]).ravel()
    
   
    training_accuracy = accuracy_score(ytrain,pred_i_training)
    training_accuracies.append(training_accuracy)
    f1_score1_training = f1_score(ytrain,pred_i_training)
    training_f1_scores.append(f1_score1_training)
    if f1_training_k < f1_score1_training: 
        maxk_f1_train = i
        f1_training_k = f1_score1_training
        
    if a_training_k<training_accuracy:
        maxk_accuracy_train = i
        a_training_k = training_accuracy
        
    #--------------------------------------------------------------------------------------------------------------
    
    knn_testing = KNeighborsClassifier(n_neighbors=i)
    knn_testing.fit(xtrain,ytrain)
    pred_i = knn_testing.predict(xtest)
    testing_error_rate.append(np.mean(pred_i !=ytest))
    
    confusion_matrix(ytest,pred_i)
    
    tn, fp, fn, tp = confusion_matrix([0, 1, 0, 1], [1, 1, 1, 0]).ravel()
    
   
    testing_accuracy = accuracy_score(ytest,pred_i)
    testing_accuracies.append(testing_accuracy)
    f1_score1_testing = f1_score(ytest,pred_i)
    testing_f1_scores.append(f1_score1_testing)
    if f1_testing_k < f1_score1_testing: 
        maxk_f1_test = i
        f1_testing_k = f1_score1_testing
        
    if a_testing_k < testing_accuracy:
        maxk_accuracy_test = i
        a_testing_k = testing_accuracy
        
    # Calculate ROC curve
    fpr, tpr, _ = roc_curve(ytest, pred_i)
    roc_auc = auc(fpr, tpr)
    roc_curve_scores.append(roc_auc)

    
print("Maximum testing F1 score is when k = ",maxk_f1_test);
print("Maximum testing F1 score is ",f1_testing_k)
print("Maximum testing Accuracy is when k = ",maxk_accuracy_test)
print("Maximum testing Accuracy is ",a_testing_k)

print("Maximum training F1 score is when k = ",maxk_f1_train)
print("Maximum training F1 score is ",f1_training_k)
print("Maximum training Accuracy is when k = ",maxk_accuracy_train)
print("Maximum training Accuracy is ",a_training_k)
```

```python
#only testing accuracies
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(range(1,60),testing_accuracies,color='blue',markersize =10,label="testing accuracy")
plt.plot(range(1,60),training_accuracies,color='green',markersize =10,label="training accuracy")
plt.title('K vs Accuracy')
plt.xlabel('k')
plt.ylabel('Accuracies')
plt.legend(['Testing Accuracy', 'Training Accuracy'])

plt.figure(figsize=(10,6))
plt.plot(range(1,60),testing_f1_scores,color='blue',markersize =10,label='testing f1 Score')
plt.plot(range(1,60),training_f1_scores,color='green',markersize =10,label = 'training f1 Score')
plt.title('K vs f1_Score')
plt.xlabel('k')
plt.ylabel('f1_scores')
plt.legend(['Testing Accuracy', 'Training Accuracy'])

plt.figure(figsize=(10,6))
plt.plot(range(1,60),roc_curve_scores,color = "blue",markersize = 10)
plt.title('K vs ROC Scores')
plt.xlabel('K')
plt.ylabel('ROC')
plt.legend(['Testing ROC'])

from sklearn.metrics import confusion_matrix, classification_report

knn = KNeighborsClassifier(n_neighbors=13)
knn.fit(xtrain, ytrain)
predictions = knn.predict(xtest)
print(confusion_matrix(ytest, predictions))
print(classification_report(ytest, predictions))
```
